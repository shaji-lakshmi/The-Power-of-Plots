## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata,study_results,on="Mouse ID")
# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
dataCount = combined_data.shape[0]
miceCount = len(pd.unique(combined_data['Mouse ID']))
print(f"Whole dataset: {dataCount}")
print(f"Unique mouse count: {miceCount}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicateMice = combined_data[combined_data.duplicated(subset=['Mouse ID'])]
duplicateMice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleanedData = combined_data.drop_duplicates(subset=['Mouse ID'], keep='last')
cleanedData

#Reference: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html#pandas.DataFrame.drop_duplicates
#Reference: https://stackoverflow.com/questions/12497402/python-pandas-remove-duplicates-by-columns-a-keeping-the-row-with-the-highest
```

```python
# Checking the number of mice in the clean DataFrame.
cleanCount = cleanedData.shape[0]
print(f"Clean data count: {cleanCount}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
myGroupBy = cleanedData.groupby('Drug Regimen')
meanVal = myGroupBy['Tumor Volume (mm3)'].mean()
medianVal = myGroupBy['Tumor Volume (mm3)'].median()
varianceVal = myGroupBy['Tumor Volume (mm3)'].var()
sdVal = myGroupBy['Tumor Volume (mm3)'].std()
semVal = myGroupBy['Tumor Volume (mm3)'].sem()

statsTable = pd.DataFrame({"Mean": meanVal,
                          "Median": medianVal,
                          "Variance": varianceVal,
                          "Standard Deviation": sdVal,
                          "SEM": semVal})
statsTable
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

summaryTable = combined_data['Drug Regimen'].value_counts()
final = pd.DataFrame(summaryTable)
final
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint


Ca = cleanedData.loc[cleanedData['Drug Regimen'] == "Capomulin"]
R = cleanedData.loc[cleanedData['Drug Regimen'] =="Ramicane"]
I = cleanedData.loc[cleanedData['Drug Regimen'] =="Infubinol"]
Ce = cleanedData.loc[cleanedData['Drug Regimen'] =="Ceftamin"]
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatmentLabels = ["Capomulin", "Ramicane", "Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumorVolCa = Ca["Tumor Volume (mm3)"]
tumorVolR = R["Tumor Volume (mm3)"]
tumorVolI= I["Tumor Volume (mm3)"]
tumorVolCe = Ce["Tumor Volume (mm3)"]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    # Determine outliers using upper and lower bounds
    
Caquart = tumorVolCa.quantile([.25,.50,.75])
CaUpperQ = Caquart[.75]
CaLowerQ = Caquart[.25]
CaIqr = CaUpperQ - CaLowerQ
LBCa = CaLowerQ - (1.5*CaIqr)
UBCa = CaUpperQ + (1.5*CaIqr)
CaOutliers=(Ca.loc[(Ca['Tumor Volume (mm3)'] < LBCa)| (Ca['Tumor Volume (mm3)']> UBCa)])
print(f"Capomulin Ourliers: {CaOutliers['Tumor Volume (mm3)']}")

Rquart = tumorVolR.quantile([.25,.50,.75])
RUpperQ = Rquart[.75]
RLowerQ = Rquart[.25]
RIqr = RUpperQ - RLowerQ
LBR = RLowerQ - (1.5*RIqr)
UBR = RUpperQ + (1.5*RIqr)

ROutliers=(R.loc[(R['Tumor Volume (mm3)'] < LBR)|(R['Tumor Volume (mm3)']> UBR)])
print(f"Ramicane Ourliers: {ROutliers['Tumor Volume (mm3)']}")

Iquart = tumorVolI.quantile([.25,.50,.75])
IUpperQ = Iquart[.75]
ILowerQ = Iquart[.25]
IIqr = IUpperQ - ILowerQ
LBI = ILowerQ - (1.5*IIqr)
UBI = IUpperQ + (1.5*IIqr)
IOutliers=(I.loc[(I['Tumor Volume (mm3)'] < LBI)|(I['Tumor Volume (mm3)']> UBI)])
print(f"Infubinol Ourliers: {IOutliers['Tumor Volume (mm3)']}")

Cequart = tumorVolCe.quantile([.25,.50,.75])
CeUpperQ = Cequart[.75]
CeLowerQ = Cequart[.25]
CeIqr = CeUpperQ - CeLowerQ
LBCe = CeLowerQ - (1.5*CeIqr)
UBCe = CeUpperQ + (1.5*CeIqr)
CeOutliers=(Ce.loc[(Ce['Tumor Volume (mm3)'] < LBCe)|(Ce['Tumor Volume (mm3)']> UBCe)])
#print(f"Ceftamin Ourliers: {CeOutliers['Tumor Volume (mm3)']}")  
    
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.boxplot([tumorVolCa,tumorVolR,tumorVolI,tumorVolCe])
ax1.set_xticklabels(treatmentLabels)
ax1.set_ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression
